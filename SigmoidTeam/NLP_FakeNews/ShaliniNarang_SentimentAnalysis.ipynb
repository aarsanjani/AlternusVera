```
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

analyser = SentimentIntensityAnalyzer()
```

```
def print_sentiment_scores(sentence):
    snt = analyser.polarity_scores(sentence)
    print(snt['neg'] )
    print("{:-<40} {}".format(sentence, str(snt)))
```

```
print_sentiment_scores("I had a pretty bad day")
```

```
##importing the file
df = pd.read_csv('https://s3-us-west-2.amazonaws.com/themodestwhite.com/fake_real_dataset_spam_category_clickbait_toxicity_politafln.txt')
df.head()
```

```
df['title'] = np.where(df.title.isnull() == True, df.text.str[:50],df.title )
```

```
new_df = df[['title','category_factor']].copy()
new_df['neg'] = np.nan
new_df['pos'] = np.nan
new_df['neu'] = np.nan
new_df['polarity'] = np.nan
new_df['sentiment_score'] = np.nan

new_df.head()
```

Finding negative,positive and neutral scores using vader library

```
def set_neg_pos_neu(df):
  for i,row in df.iterrows():
    title=row['title']
    snt = analyser.polarity_scores(title)
    df.loc[i,'neg'] = snt['neg']
    df.loc[i,'pos'] = snt['pos']
    df.loc[i,'neu'] = snt['neu']
```

```
set_neg_pos_neu(new_df)
new_df.head()
```

Finding polarity using TextBlob

```
def set_polarity(df):
  for i,row in df.iterrows():
    title=row['title']
    opinion = TextBlob(title)
    df.loc[i,'polarity'] = opinion.sentiment.polarity
    
```

```
set_polarity(new_df)
new_df.head()
```
